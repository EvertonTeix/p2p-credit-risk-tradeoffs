```python
numerical_cols=[col for col in database.columns if database[col].dtype!=object]
categorical_cols=[col for col in database.columns if database[col].dtype==object]
print("No of Numerical features =",len(numerical_cols))
print("No of Categorical features =",len(categorical_cols))
```

```python
def Numerical_Summary():
    for col in numerical_cols:
        print(col)
        print("No of Missing Values =",database[col].isnull().sum())
        print("Skew =",database[col].skew())
        print("Histogram")
        fig=px.histogram(data_frame=database,x=col)
        fig.show()
        print("="*50)

def Categorical_Summary():
    for col in categorical_cols:
        print(col)
        print("No of Missin Values =",database[col].isnull().sum())
        print(database[col].value_counts())
        print("="*50)
```
